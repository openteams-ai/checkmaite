## Perturbers

```python
# specify the root visdrone directory
from pathlib import Path
from jatic_ri import PACKAGE_DIR

ROOT_VISDRONE_DIR = PACKAGE_DIR.parents[1] / "tests/testing_utilities/example_data/visdrone_dataset"
# Use cv2 & matplotlib to visualize the perturbers.
import numpy as np
import matplotlib.pyplot as plt
from torchvision.transforms.functional import to_pil_image
from jatic_ri.object_detection.datasets import VisdroneDetectionDataset
```

```python
# Display images with matplotlib
def display_orig_and_pert_image(original_image, image, title):
    fig, ax = plt.subplots(1, 2, figsize=(20, 20))
    ax[0].imshow(original_image)
    ax[0].set_title('Original Image')
    ax[0].set_axis_off()
    ax[1].imshow(image)
    ax[1].set_title(title)
    ax[1].set_axis_off()
    plt.tight_layout()
    plt.show()

# Load the VisDrone dataset
vd = VisdroneDetectionDataset(ROOT_VISDRONE_DIR)

# Get the first image and its annotations
img, anns, _ = vd[0]
image_to_use = np.asarray(to_pil_image(img))
```

### Average Blur Perturber

The average blur perturber uses Open CV to apply an "average" blurring over the image. This perturber has a single variable [ksize](#ksize). The larger the ksize the more the blur effect on the image. ksize must be greater than or equal to 1.

[Back to Table of Contents](#table-of-contents)

```python
# Import Average blur from nrtk
from nrtk.impls.perturb_image.generic.cv2.blur import AverageBlurPerturber
```

```python
# Set up the perturber 
ksize = 1
perturber = AverageBlurPerturber(
    ksize=ksize
)
# Perturb the image using the perturber
perturbed_img = perturber(image_to_use)[0]
# Display the perturbed image
display_orig_and_pert_image(image_to_use, perturbed_img, f"Perturbed Image (Average Blur Ksize: {ksize})")
```

```python
# Set up the perturber 
ksize = 5
perturber = AverageBlurPerturber(
    ksize=ksize
)
# Perturb the image using the perturber
perturbed_img = perturber(image_to_use)[0]
# Display the perturbed image
display_orig_and_pert_image(image_to_use, perturbed_img, f"Perturbed Image (Average Blur Ksize: {ksize})")
```

```python
# Set up the perturber 
ksize = 10
perturber = AverageBlurPerturber(
    ksize=ksize
)
# Perturb the image using the perturber
perturbed_img = perturber(image_to_use)[0]
# Display the perturbed image
display_orig_and_pert_image(image_to_use, perturbed_img, f"Perturbed Image (Average Blur Ksize: {ksize})")
```

### Brightness Perturber

The brightness perturber only has 1 available variable, which is [Factor](#factor).
Factor can be considered a percentage for how bright you want to make the image. 
A factor value of 1.0 is equivalent to 100% brightness or the original brightness value. 2.0 would be 200% brightness making the image very bright and 0.5 would be 50% which would make the image darker.

[Back to Table of Contents](#table-of-contents)

### Pybsm Perturber

Pybsm is the most complex of the pertubers in NRTK. It uses a PybsmSensor & PybsmScenario to set the scope of how the image was orinally taken. Then by altering the values you can perturb the image to show how it should have looked if the factors had been different. To run a single instance of the perturber you would need to set up the Sensor, Scenario, and then pick which variable you would want to change.

Variable Lists:
- Scenario
   - [ihaze](#ihaze)
   - [altitude](#altitude)
   - [ground_range](#ground_range)
   - [aircraft_speed](#aircraft_speed)
- Sensor
   - [D](#D)
   - [f](#f)
   - [p](#p)
   - [opt_trans_wavelengths](#opt_trans_wavelengths)
   - [optics_transmission](#optics_transmission)
   - [eta](#eta)
   - [w_x](#w_x)
   - [w_y](#w_y)
   - [int_time](#int_time)
   - [n_tdi](#n_tdi)
   - [dark_current](#dark_current)
   - [read_noise](#read_noise)
   - [max_n](#max_n)
   - [bit_depth](#bit_depth)
   - [max_well_fill](#max_well_fill)

   - [s_y](#s_y)

   - [da_y](#da_y)
   - [qe_wavelengths](#qe_wavelengths)
   - [qe](#qe)

[Back to Table of Contents](#table-of-contents)

```python
# Import PybsmPerturber from nrtk, as well as PybsmSensor and PybsmScenario that are needed for the PybsmPerturber.
from nrtk.impls.perturb_image.pybsm.scenario import PybsmScenario
from nrtk.impls.perturb_image.pybsm.sensor import PybsmSensor
from nrtk.impls.perturb_image.pybsm.perturber import PybsmPerturber
```

```python
from pybsm.otf import dark_current_from_density

# Set up Sensor
name = "TestSensor"

# telescope focal length (m)
f = 4
# Telescope diameter (m)
D = 275e-3

# detector pitch (m)
p = 0.008e-3

# Optical system transmission, red  band first (m)
opt_trans_wavelengths = np.array([0.58 - 0.08, 0.58 + 0.08]) * 1.0e-6
# guess at the full system optical transmission (excluding obscuration)
optics_transmission = 0.5 * np.ones(opt_trans_wavelengths.shape[0])

# Relative linear telescope obscuration
eta = 0.4  # guess

# detector width is assumed to be equal to the pitch
w_x = p
w_y = p
# integration time (s) - this is a maximum, the actual integration time will be
# determined by the well fill percentage
int_time = 30.0e-3

# the number of time-delay integration stages (relevant only when TDI
# cameras are used. For CMOS cameras, the value can be assumed to be 1.0)
n_tdi = 1.0

# dark current density of 1 nA/cm2 guess, guess mid range for a silicon camera
dark_current = dark_current_from_density(1e-5, w_x, w_y)

# rms read noise (rms electrons)
read_noise = 25.0

# maximum ADC level (electrons)
max_n = 96000

# bit depth
bit_depth = 11.9

# maximum allowable well fill (see the paper for the logic behind this)
max_well_fill = 0.6

# jitter (radians) - The Olson paper says that its "good" so we'll guess 1/4 ifov rms
s_x = 0.25 * p / f
s_y = s_x

# drift (radians/s) - again, we'll guess that it's really good
da_x = 100e-6
da_y = da_x

# etector quantum efficiency as a function of wavelength (microns)
# for a generic high quality back-illuminated silicon array
# https://www.photometrics.com/resources/learningzone/quantumefficiency.php
qe_wavelengths = np.array([0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1]) * 1.0e-6
qe = np.array([0.05, 0.6, 0.75, 0.85, 0.85, 0.75, 0.5, 0.2, 0])

sensor = PybsmSensor(
    name,
    D,
    f,
    p,
    opt_trans_wavelengths,
    optics_transmission,
    eta,
    w_x,
    w_y,
    int_time,
    n_tdi,
    dark_current,
    read_noise,
    max_n,
    bit_depth,
    max_well_fill,
    s_x,
    s_y,
    da_x,
    da_y,
    qe_wavelengths,
    qe,
)
```

```python
# Set up the Scenario

scenario_name = "TestScenario"

# sensor altitude
altitude = 9000.0

# range to target
ground_range = 0.0

# weather model
ihaze = 1

# ground speed in m/s
aircraft_speed = 100.0  

scenario = PybsmScenario(scenario_name, ihaze, altitude, ground_range, aircraft_speed)
```

```python
# Set up the perturber 
img_gsd = 3.19 / 165.0  # the width of the tank is 319 cm and it spans ~165 pixels in the image
perturber = PybsmPerturber(sensor=sensor, scenario=scenario, ground_range=10000)
# Perturb the image using the perturber
perturbed_img = perturber(image_to_use, additional_params={"img_gsd": img_gsd})[0]
# Display the perturbed image
display_orig_and_pert_image(image_to_use, perturbed_img, f"Perturbed Image (PybsmPerturber Ground Range: 10000)")
```

```python
# Set up the perturber 
img_gsd = 3.19 / 165.0  # the width of the tank is 319 cm and it spans ~165 pixels in the image
perturber = PybsmPerturber(sensor=sensor, scenario=scenario, altitude=20000)
# Perturb the image using the perturber
perturbed_img = perturber(image_to_use, additional_params={"img_gsd": img_gsd})[0]
# Display the perturbed image
display_orig_and_pert_image(image_to_use, perturbed_img, f"Perturbed Image (PybsmPerturber Altitude: 20000)")
```

## Factory Types

If you think of a perturber as an object that can alter images based on a set of rules. Then you can think of the image factories as an object that can produce mulitple perturbers each with a slightly different set of rules.

[Back to Table of Contents](#table-of-contents)

### LinSpacePerturbImageFactory

This factory will produce multiple perturbers based on a theta key (variable), start (initial value), stop (value not included, so the factory will stop prior to reaching the stop value), and step (how many perturbers will be made.)

- Example - For this we can again consider the brightness perturber, having the theta_key be factor and the thetas be 0.5, 1.0, 1.5, and 2.0 there would be a total of 4 brightness perturbers being created. With the Jatic-ri you would define this using start (0.5), stop (2.5), & step (4).
    - Start - is the first theta value being used, this value will be included in the final set.
    - Stop - this is the stopping point, this value will not be included in the final set.
    - Step - how many total values will be created.
    - Value increment (i) calculation is essentially (Stop-Start)/Step.
        - In the above example (2.5 - 0.5) / 4 = 0.5
        - The values then become [start, start+i, ...], in this case [0.5, 1.0, 1.5, 2.0]


[Back to Table of Contents](#table-of-contents)

### OneStepPerturbImageFactory

This is the simplest of the perturber factory types, with this one you select your perturber, the theta key (variable) you wish to use, and the theta (value) that will be used for the perturbment.

[Back to Table of Contents](#table-of-contents)

### StepPerturbImageFactory

The step perturber factory type is very similiar to the LinSpacePerturbImageFactory with the theta_key, start, stop, & step variables. However, this factory also has the addition of an optional variable to_int which is a boolean flag. The big difference is that in the LinSpacePerturbImageFactory step referes to the number of steps, while in the StepPerturbImageFactory it is the size of each step. 

The formulas for this factory are:
- to_int = False:
   - [start + i * step for i in range(math.ceil((stop - start) / step))]
- to_int = True:
   - [int(start + i * step) for i in range(math.ceil((stop - start) / step))]

Here is an example:
start = 0
stop = 10
step = 2.5

- to_int = False:
   1) [0 + i * 2.5 for i in range(math.ceil((10 - 0) / 2.5))]
   2) [0 + i * 2.5 for i in range(math.ceil(10 / 2.5))]
   3) [0 + i * 2.5 for i in range(math.ceil(4))]
   4) [0 + i * 2.5 for i in range(4)]
   5) [0 + i * 2.5 for i in range(0, 4)]
        - 0 + 0 * 2.5 = 0
        - 0 + 1 * 2.5 = 2.5
        - 0 + 2 * 2.5 = 5
        - 0 + 3 * 2.5 = 7.5
   6) [0, 2.5, 5, 7.5]
- to_int = True:
   1) [int(0 + i * 2.5) for i in range(math.ceil((10 - 0) / 2.5))]
   2) [int(0 + i * 2.5) for i in range(math.ceil(10 / 2.5))]
   3) [int(0 + i * 2.5) for i in range(math.ceil(4))]
   4) [int(0 + i * 2.5) for i in range(4)]
   5) [int(0 + i * 2.5) for i in range(0, 4)]
        - int(0 + 0 * 2.5) = 0
        - int(0 + 1 * 2.5) = 2
        - int(0 + 2 * 2.5) = 5
        - int(0 + 3 * 2.5) = 7
   6) [0, 2, 5, 7]
