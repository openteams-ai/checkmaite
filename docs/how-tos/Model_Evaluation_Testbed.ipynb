# Model Evaluation Testbed

The Model Evaluation Testbed is an app designed to run analyses 

The first step is to use the Model Evaluation Configuration app to generate your *.json* configuration file. 
Once you have the *.json* configuration, load it into this interface and select the datasets and models you'd
like to analyze along with a metric to run. You also have the option of specifying a target threshold which
will be used for visualizations. 

This app also contains a cacheing mechanism which allows you to cache predictions and metrics so that 
subsequent analyses are computed much faster. 

Once the analyses are complete, the results will appear in the `Test Results` table with a link to a `*.ppt` 
report summarizing the results.

The Model Evaluation Testbed supports the following for both   
**Image Classfication** and **Object Detection**:

* NRTK
* XAITK
* Baseline Evaluation (maite)

*NOTE: Viewing is optimized on monitors which allow 1280 px width for the app.*

## Setup

* Define what task you are currently targeting - **Object Detection** or **Image Classification**
* Define the output directory for the report
* Set `local=True` for local deployments, or `local=False` for remote server deployments.
  * For server deployments, users are presented a button to download Gradient Report while on local deployments the filepath is displayed. 

```python
import os

# task = 'image_classification'
task = 'object_detection'
local = True

report_output_dir = '.'
```

## Instantiate and Run the app

```python
from jatic_ri._common._panel.dashboards.model_evaluation_dashboard import ModelEvaluationTestbed

# instantiate the app
app = ModelEvaluationTestbed(
    task=task,
    local=local,
    output_dir=report_output_dir,
)
app.panel().servable()
```
