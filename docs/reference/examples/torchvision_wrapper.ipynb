## What is `torchvision`?

`torchvision` is a library in the PyTorch ecosystem that provides tools specifically for computer vision tasks. It includes:

- Pretrained Models: Access to a collection of popular pretrained models (e.g., for image classification and object detection) that can be used for evaluation.
- Transforms: A set of utilities to preprocess and augment images, such as resizing, cropping, normalizing, and converting images to tensors. This is useful for preparing data in a way that meets model requirements.

## Why use `torchvision` in AI model testing and evaluation?
`torchvision` offers:

- Quick Start with Pretrained Models: You can load and test models trained on large datasets (like ImageNet) without having to train from scratch.
- Preprocessing Consistency: `torchvision`’s transforms help ensure that all images are processed consistently, which is essential for evaluating models reliably.

## What does the `torchvision` wrapper do?

It provides a MAITE-compliant wrapper for running `torchvision` models. This notebook will give a brief introduction on how to do this.

## How to use the `torchvision` wrapper

```python
from jatic_ri.object_detection.models import TorchvisionODModel
import numpy as np
import torch
from PIL import Image
```

```python
# example image for testing
img = Image.open("../../tests/testing_utilities/example_data/coco_dataset/000000037777.jpg")

# PIL loads data as HWC, but MAITE requires CHW
img_chw = torch.tensor(np.array(img).transpose(2, 0, 1))

img.show()
```

```python
# see src/jatic_ri/object_detection/models.py current list of supported models
torch_wrapper = TorchvisionODModel(model_name="fasterrcnn_resnet50_fpn_v2", device="cpu")
```

```python
# MAITE expects batch input (N, C, H, W) of type Sequence[ArrayLike]
# here we create a batch of size N=1
batch = [img_chw]
prediction = torch_wrapper(input_batch=batch)
```

```python
# we have a single prediction as there is only one model and one image
print(f"Our model prediction has {len(prediction[0].boxes)} bounding boxes.")
```

```python
from torchvision.utils import draw_bounding_boxes
from torchvision.io.image import decode_image
from torchvision.transforms.functional import to_pil_image

# only displaying the first 5 bounding boxes, unreadable otherwise
box = draw_bounding_boxes(img_chw, boxes=prediction[0].boxes[0:5], labels=[torch_wrapper.index2label[i] for i in prediction[0].labels[0:5]], colors="red", width=4)
im = to_pil_image(box)
im.show()
```
