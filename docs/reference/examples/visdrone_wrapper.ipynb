## What is Visdrone?

Visdrone is a dataset for drone based detection and tracking. See https://github.com/VisDrone/VisDrone-Dataset for details.

## What does the Visdrone wrapper do?

It provides a MAITE-compliant wrapper around [models](https://github.com/GNAYUOHZ/centernet-visdrone?tab=readme-ov-file) that have been trained on the Visdrone dataset, and have been [packaged](https://github.com/Kitware/SMQTK-Detection/blob/master/smqtk_detection/impls/detect_image_objects/centernet.py) for distribution by Kitware.

## How to use the Visdrone wrapper

```python
from jatic_ri.object_detection.models import VisdroneODModel
import torch
from PIL import Image
import numpy as np
```

```python
# example image for testing
img = Image.open("../../tests/testing_utilities/example_data/coco_dataset/000000087038.jpg")

# PIL loads data as HWC, but MAITE requires CHW
img_chw = torch.tensor(np.array(img).transpose(2, 0, 1))

img.show()
```

```python
# model_pickle_dir is where the pickle will be downloaded and saved to
visdrone_wrapper = VisdroneODModel(arch="resnet50", model_pickle_dir="visdrone")
```

```python
batch = [img_chw]
prediction=visdrone_wrapper(input_batch=batch)
```

```python
# we have a single prediction as there is only one model and one image
print(f"Our model prediction has {len(prediction[0].boxes)} bounding boxes.")
```

```python
from torchvision.utils import draw_bounding_boxes
from torchvision.io.image import decode_image
from torchvision.transforms.functional import to_pil_image

# only displaying the first 10 bounding boxes, unreadable otherwise
box = draw_bounding_boxes(img_chw, boxes=torch.tensor(prediction[0].boxes[0:10]), labels=[visdrone_wrapper.index2label[i] for i in prediction[0].labels[0:10]], colors="red", width=4)
im = to_pil_image(box)
im.show()
```
