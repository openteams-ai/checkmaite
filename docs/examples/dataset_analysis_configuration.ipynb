# Dataset Analysis Configuration App - Object Detection

A quick guide to the Dataset Analysis configuration app for object detection

```python
from jatic_ri.object_detection._panel.configurations.dataset_analysis_configuration import DatasetAnalysisConfigApp
```

## Instantiate and view the app

The addition of `.servable()` here allows us also to deploy this app via this notebook using `panel serve dataset_analysis_configuration.ipynb`

```python
app = DatasetAnalysisConfigApp(task='object_detection')
app.panel().servable()
```

The `dataeval` configurations are simple boolean toggles on the front page. `Survivor` and `Reallabel` are separate configuration pages. Triggering the toggles on the front page will add/remove the followup config pages. 

## Investigating the UI programmatically

The visualized UI objects can be accessed programmatically, which is helpful for debugging purposes. 

```python
# view the name of the page that's currently visualized
app.pipeline._stage
```

```python
# access the class object of the page that's currently visualized
# this can be used to access attributes of the class (widgets, etc)
page_class = app.pipeline._state
page_class
```

```python
# programmatically move between stages
app.pipeline.next_button.clicks += 1
```

```python
# programmatically move between stages
app.pipeline.prev_button.clicks += 1
```
