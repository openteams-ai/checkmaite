## What do dataset wrappers do?

They provide a MAITE-compliant wrapper for loading computer-vision image classification datasets. This notebook will give a brief introduction on how to do this for various wrappers

### How to use the YOLO image classification dataset wrapper

The YOLO image classification dataset wrapper loads computer vision datasets that are available on a local machine and are laid out according to the YOLO specification. See the 
[official documentation](https://docs.ultralytics.com/datasets/classify/) for more details.

The sample dataset in this notebook is created and stored on disk in the following way:
```
my-dataset-yolo-layout-/
|
|-- test/
|   |-- cat/
|   |   |-- 1_cat.jpg
|   |   |-- 2_cat.jpg
|   |
|   |-- dog/
|   |   |-- 1_dog.jpg
|   |   |-- 2_dog.jpg
|   |
|
```

````python
import os
from pathlib import Path

from PIL import Image

DATA_DIR = "my-dataset-yolo-layout"

# Warning: this will create a fake dataset on your filesystem at ./DATA_DIR
def create_fake_yolo_dataset(root_dir, split, classes, num_images_per_class):
    """
    Creates a fake dataset structure resembling a YOLO-style dataset.

    This function generates a directory structure where each class has its own folder
    under a specific split (e.g., "train" or "test"). Inside each class folder, it creates
    a specified number of fake images with simple grayscale patterns.

    Args:
        root_dir (Path or str): 
            The root directory where the dataset will be created.
        split (str): 
            The dataset split (e.g., "train", "validation", "test") under which the class folders
            and images will be stored.
        classes (list of str): 
            A list of class names to create directories for.
        num_images_per_class (int): 
            The number of fake images to generate for each class.
    """
    os.makedirs(root_dir / split, exist_ok=True)
    for class_name in classes:
        class_dir = root_dir / split / class_name
        os.makedirs(class_dir, exist_ok=True)
        for i in range(num_images_per_class):
            img = Image.new("RGB", (64, 128), color=(i, i, i))
            img.save(class_dir / f"{i}_{class_name}.jpg")

create_fake_yolo_dataset(
    root_dir=Path(DATA_DIR),
    split="test",
    classes=["cat", "dog"],
    num_images_per_class=2,
)

os.listdir(DATA_DIR)
````
