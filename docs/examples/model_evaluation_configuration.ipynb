# Model Evaluation Configuration App

A configuration wizard for Object Detection and Image Classification usecases

**Tools included:**
* NRTK
* XAITK
* maite (as baseline evaluation)

```python
from jatic_ri._common._panel.configurations.model_evaluation_configuration import ModelEvaluationConfigApp
```

## Define usecase

* Set the task for the appropriate usecase, either `"object_detection"` or `"image_classification"`  
* Set `local=True` for local deployments, or `local=False` for remote server deployments.
  * For server deployments, users are presented a button to download results while on local deployments the filepath is displayed. 

```python
task = 'image_classification'
local = True
```

## Instantiate and view the app

The addition of `.servable()` here allows us also to deploy this app via this notebook using `panel serve model_evaluation_configuration.ipynb`

```python
app = ModelEvaluationConfigApp(task=task, local=local)
app.panel().servable()
```

The `Baseline Evaluation` (aka `maite`) configuration is a simple boolean toggles on the front page. `NRTK` and `XAITK` are separate configuration pages. Triggering the toggles on the front page will add/remove the followup config pages. 

## Investigating the UI programmatically

The visualized UI objects can be accessed programmatically, which is helpful for debugging purposes. 

```python
# view the name of the page that's currently visualized
app.pipeline._stage
```

```python
# access the class object of the page that's currently visualized
# this can be used to access attributes of the class (widgets, etc)
page_class = app.pipeline._state
page_class
```

```python
# view the output configuration that will be downloaded
page_class.output_test_stages
```

```python
# programmatically move between stages (as long as the button isn't disabled)
if not app.pipeline.next_button.disabled:
    app.pipeline.next_button.clicks += 1
else:
    print('There is no next stage.')
```

```python
# programmatically move between stages (as long as the button isn't disabled)
if not app.pipeline.prev_button.disabled:
    app.pipeline.prev_button.clicks += 1
else: 
    print('There is no previous stage.')
```
